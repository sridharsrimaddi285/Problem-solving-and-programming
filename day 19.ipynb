## PROBLEM 2

## PROBLEM STATEMENT:
Define a function to add new row to data to a csv file


```python
def addrowcsv(filename,rowdata):
    df = readCSV(filename)
    df.loc[len(df)] = rowdata
    df.to_csv(filename)
    return df

rowdata = ['010101','statename',123,45,654,345,564,987,987,888,666]
addrowcsv('csv.txt',rowdata)
```

```python
def updaterowcsv(filename,rowindex,rowdata):
    df = readCSV(filename)
    df.loc[rowindex] = rowdata
    df.to_csv(filename,index = False)
    return df

rowdata = ['04000US08','Florida',123,234,456,765,897,567,345,765,345]
updaterowcsv('csv.txt',6,rowdata)

```

## PROBLEM 3

## PROBLEM STATEMENT:

Define a function to delete a row from csv


```python
def deleterowcsv(filename,rowindex):
    df = readCSV(filename)
    df = df.drop(rowindex)
    df.to_csv(filename,index = False)
    return df
deleterowcsv('csv.txt',6)
```
