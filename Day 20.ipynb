```python
import pandas as pd
df = pd.read_csv('csv.txt')
y= df.values[4]
x = df.columns
years = []
for i in range(2,len(x)):
    years.append(x[i])
years
income = []
for i in range(2,len(y)):
    income.append(y[i])
income
plt.xlabel('year')
plt.ylabel('income')
plt.plot(years,income)
plt.grid(True)
plt.savefig('figures/pictures1')
plt.show()
```

```python
width = 1/2
plt.bar(years,income,width,color='grey')
plt.savefig('figures/pictures2')
```

```python
N = 50

x = np.random.rand(N)
y = np.random.rand(N)
colorlist = np.random.rand(N)
arealist = np.pi*(15*np.random.rand(50))**2
plt.scatter(x,y,s=arealist,c=colorlist,alpha=1)
plt.savefig('figures/pictures3')
plt.show()
```
